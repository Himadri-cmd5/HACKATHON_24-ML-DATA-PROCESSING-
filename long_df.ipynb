```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df_1 = pd.read_csv("D:\\HACKHERITAGE\\primary_dataset.csv")
```

```python
df_1
```

```python
df_1['BARLEY HARVEST PRICE (Rs per Quintal)'] = np.where(df_1['BARLEY HARVEST PRICE (Rs per Quintal)'] == -1, np.nan, df_1['BARLEY HARVEST PRICE (Rs per Quintal)'])
df_1['FINGER MILLET HARVEST PRICE (Rs per Quintal)'] = np.where(df_1['FINGER MILLET HARVEST PRICE (Rs per Quintal)'] == -1, np.nan, df_1['FINGER MILLET HARVEST PRICE (Rs per Quintal)'])
df_1['MAIZE HARVEST PRICE (Rs per Quintal)'] = np.where(df_1['MAIZE HARVEST PRICE (Rs per Quintal)'] == -1, np.nan, df_1['MAIZE HARVEST PRICE (Rs per Quintal)'])
df_1['SORGHUM HARVEST PRICE (Rs per Quintal)'] = np.where(df_1['SORGHUM HARVEST PRICE (Rs per Quintal)'] == -1, np.nan, df_1['SORGHUM HARVEST PRICE (Rs per Quintal)'])    
df_1['PADDY HARVEST PRICE (Rs per Quintal)'] = np.where(df_1['PADDY HARVEST PRICE (Rs per Quintal)'] == -1, np.nan, df_1['PADDY HARVEST PRICE (Rs per Quintal)'])
df_1['WHEAT HARVEST PRICE (Rs per Quintal)'] = np.where(df_1['WHEAT HARVEST PRICE (Rs per Quintal)'] == -1, np.nan, df_1['WHEAT HARVEST PRICE (Rs per Quintal)'])                                                     
```

```python
df_1['BARLEY HARVEST PRICE (Rs per Quintal)'] = np.where(df_1['BARLEY HARVEST PRICE (Rs per Quintal)'].isna(), df_1['BARLEY HARVEST PRICE (Rs per Quintal)'].mean(), df_1['BARLEY HARVEST PRICE (Rs per Quintal)'])
df_1['FINGER MILLET HARVEST PRICE (Rs per Quintal)'] = np.where(df_1['FINGER MILLET HARVEST PRICE (Rs per Quintal)'].isna(), df_1['FINGER MILLET HARVEST PRICE (Rs per Quintal)'].mean(), df_1['FINGER MILLET HARVEST PRICE (Rs per Quintal)'])
df_1['MAIZE HARVEST PRICE (Rs per Quintal)'] = np.where(df_1['MAIZE HARVEST PRICE (Rs per Quintal)'].isna(), df_1['MAIZE HARVEST PRICE (Rs per Quintal)'].mean(), df_1['MAIZE HARVEST PRICE (Rs per Quintal)'])
df_1['SORGHUM HARVEST PRICE (Rs per Quintal)'] = np.where(df_1['SORGHUM HARVEST PRICE (Rs per Quintal)'].isna(), df_1['SORGHUM HARVEST PRICE (Rs per Quintal)'].mean(), df_1['SORGHUM HARVEST PRICE (Rs per Quintal)'])    
df_1['PADDY HARVEST PRICE (Rs per Quintal)'] = np.where(df_1['PADDY HARVEST PRICE (Rs per Quintal)'].isna(), df_1['PADDY HARVEST PRICE (Rs per Quintal)'].mean(), df_1['PADDY HARVEST PRICE (Rs per Quintal)'])
df_1['WHEAT HARVEST PRICE (Rs per Quintal)'] = np.where(df_1['WHEAT HARVEST PRICE (Rs per Quintal)'].isna(), df_1['WHEAT HARVEST PRICE (Rs per Quintal)'], df_1['WHEAT HARVEST PRICE (Rs per Quintal)']) 
```

```python
df_1
```

```python
df = df_1.rename(columns = {'BARLEY HARVEST PRICE (Rs per Quintal)' : 'Barley'})
df = df.rename(columns = {'FINGER MILLET HARVEST PRICE (Rs per Quintal)' : 'Finger Millet'})
df = df.rename(columns = {'MAIZE HARVEST PRICE (Rs per Quintal)' : 'Maize'})
df = df.rename(columns = {'SORGHUM HARVEST PRICE (Rs per Quintal)' : 'Sorghum'})
df = df.rename(columns = {'PADDY HARVEST PRICE (Rs per Quintal)' : 'Paddy'})
df = df.rename(columns = {'WHEAT HARVEST PRICE (Rs per Quintal)' : 'Wheat'})
```

```python
df
```

```python
long_df = pd.melt(df, id_vars=['State Name', 'Dist Name', 'Year'], var_name='Crop Name', value_name='Price (Rs. per Quintal)')
```

```python
long_df
```

```python
long_df['Price (Rs. per Quintal)'] = np.where(long_df['Price (Rs. per Quintal)'].isna(), long_df['Price (Rs. per Quintal)'].mean(), long_df['Price (Rs. per Quintal)'])
```

```python
long_df.to_csv("D:\\HACKHERITAGE\\5_CROPS_1966-2016_ALL_STATES.csv", index = False)
```
